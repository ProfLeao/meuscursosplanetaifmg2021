# Plotagem

## Fundamentos
Existem algumas maneiras diferentes de plotar em Julia (incluindo o comando PyPlot). <br>

Aqui, mostraremos como usar o comando `Plots.jl`.  Se ainda não estiver instalado, você precisará usar o gerenciador de pacotes para instalá-lo e Julia irá pré-compilá-lo para você na primeira vez que usá-lo:

````julia
#Usando o Pacote
#Adicionando o pacote("Plots")
using Plots
````

Uma das vantagens do `Plots.jl` é que ele permite que você altere os backends sem problemas. Neste  notebook, vamos experimentar o `gr()` e `plotlyjs()` backends.<br>

Em nome da investigação científica, vamos usar este Notebook para examinar a relação entre a temperatura global e o número de piratas aproximado, entre 1860 e 2000.

````julia
temperaturaglobal = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpiratas = [45000, 20000, 15000, 5000, 400, 17]
````

Plots suportam múltiplos backends — ou seja, bibliotecas que realmente fazem o desenho — todos com a mesma API. Para começar, vamos tentar o backends GR.  Você escolhe com um comando para `gr()`:

````julia
gr()
````

e agora podemos usar comandos como `plot` e `scatter` para gerar plotagens.

````julia
plot(numpiratas, temperaturaglobal, label="line")  
scatter!(numpiratas, temperaturaglobal, label="points") 
````

O `!` no final do `scatter!` nome da função faz `scatter!` uma função mutante, indicando que os pontos espalhados serão adicionados ao gráfico pré-existente.

Em contraste, veja o que acontece quando você substitui `scatter!` acima com a função não mutante `scatter`.

A seguir, vamos atualizar este gráfico com o `xlabel!`, `ylabel!`, e `title!` comandos para adicionar mais informações ao nosso gráfico.

````julia
xlabel!("Número de Piratas [Aproximado]")
ylabel!("Temperatura Global (C)")
title!("Influência da população pirata no aquecimento global")
````

Isso ainda não parece certo. O número de piratas diminuiu desde 1860, portanto, ler o enredo da esquerda para a direita é como olhar para trás no tempo, em vez de para a frente. Vamos inverter o eixo x para ver melhor como as populações piratas fizeram com que as temperaturas globais mudassem ao longo do tempo!

````julia
xflip!()
````

E aí temos que!

Nota: Tivemos alguma confusão sobre este exercício. :) Esta é uma piada sobre como as pessoas costumam confundir correlação e causalidade.

**Sem alterar a sintaxe, podemos criar este gráfico com o backend UnicodePlots**

````julia
#Adicionando o Pacote ("UnicodePlots")
unicodeplots()
````

````julia
plot(numpiratas, temperaturaglobal, label="line")  
scatter!(numpiratas, temperaturaglobal, label="points") 
xlabel!("Número de Piratas [Aproximado]")
ylabel!("Temperatura Global  (C)")
title!("Influência da população pirata no aquecimento global")
````

E observe como este segundo gráfico difere do primeiro! Usar texto como este é um pouco bobo em um notebook Jupyter, onde temos recursos de desenho sofisticados, mas pode ser muito útil para uma visualização rápida e suja em um terminal.
